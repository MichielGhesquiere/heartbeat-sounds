```python
import pandas as pd 
import librosa as lr
from glob import glob
import numpy as np
```

```python
import pandas as pd 

DATA_DIR = '../data/'

set_a = pd.read_csv(DATA_DIR+'set_a.csv')
set_a.head()
```

```python
audio_files = glob(DATA_DIR + '**/*.wav', recursive=True)
audio_files
```

```python
audios = []
sfreqs = []
for f in audio_files:
    audio, sfreq = lr.load(f)
    audios.append(audio)
    sfreqs.append(sfreq)

print(f"Length audio {audios[0].shape}")
print(f"Nr of recordings: {len(audios)}")
sfreqs = np.array(sfreqs)
print(f"sfreqs mean: {sfreqs.mean()}, min: {sfreqs.min()}, max: {sfreqs.max()}")
```

```python
import matplotlib.pyplot as plt

time_0 = np.arange(0, len(audios[0])) / sfreqs[0]

fig, ax = plt.subplots()
ax.plot(time_0, audios[0])
ax.set(xlabel='Time (s)', ylabel='Signal Amplitude')
plt.show()
```
